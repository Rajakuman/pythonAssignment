## **set**
Add

```
fruits = {"apple", "banana", "cherry"}

fruits.add("orange")

print(fruits)
```

Copy

```

x=fruits.copy()

print(x)
```

difference

```
x = {"apple", "banana", "cherry"}
y = {"google", "microsoft", "apple"}

z = x.difference(y)

print(z)
```

difference_update

```
x.difference_update(y)
print(x)
```

update

```
x.update(y)
print(x)
```

discard

```
print(fruits)
fruits.discard("banana")
print(fruits)
```

intersection


```
x = {"apple", "banana", "cherry"}
y = {"google", "microsoft", "apple"}
z = x.intersection(y)

print(z)
```

union

```
z = x.union(y)
print(z)
```
