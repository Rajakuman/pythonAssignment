Question 1:
Write a program to identify sub list [1,1,5] is there in the given list in the same order,

if yes print “it's a Match” if no then print “it's Gone” in function

Example -

Listy = [1,5,6,4,1,2,3,5] - it's a Match

Listy = [1,5,6,5,1,2,3,6] - it's Gone


```
input_values=input("List of Number")
input_sub_values=input("List of sub Number")
lst_sb_values=input_sub_values.split(" ")
checkpoint=0;
b=True
print("Input values:",input_values)
print("Input sub values:",input_sub_values)
for it in lst_sb_values:
  checkpoint=input_values.find(it,checkpoint)
  if b and checkpoint==-1:
    b=False
    break


if b:
  print("it's a Match")
else:
  print("it's Gone")
```

# **Question 2:**
ake a calculator using Functions, along with If, elif and else

**Addition, Subtraction, Multiplication, Division**

You enter **two numbers and operation** you want to do, it will return the answer

```
def calculator(num1,num2,str_opr):
  num3=0
  if str_opr=="+":
    num3=num1+num2
  elif str_opr=="-":
    num3=num1-num2
  elif str_opr=="*":
    num3=num1*num2
  elif str_opr=="/":
    num3=num1/num2
  return num3

print("Enter the two number between space and with oprator(+,-,*,/)")
input_cal_values=input("Enter the values:  ")
lst_input_cal_values=input_cal_values.split(" ")
print(lst_input_cal_values[0],lst_input_cal_values[2],lst_input_cal_values[1], " = ",calculator(int(lst_input_cal_values[0]),int(lst_input_cal_values[1]),lst_input_cal_values[2]))
```
