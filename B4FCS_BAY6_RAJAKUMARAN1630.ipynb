For this challenge, create a Cone class that has two attributes:
* R = Radius
* h = height
and two methods:
* Volume - π × r² × (h/3)
* Surface area: base: π × r², side: π × r × √(r²+h²)
Make only one class with functions, as in where required Import Math

Referrence link:
https://byjus.com/maths/surface-areas-volumes/#:~:text=Surface%20Areas%20and%20Volume,space%20available%20in%20an%20object.

l = Length

h = height

r = radius

l = Squre of **√(h^2 + r^2)**

formula:

Total Surface Area	: **A=πr(r+ √(h^2+r^2))**

Curved Surface Area : **π r l**

Volume  :  **πr^2(h/3)**


```
import math

class cone:
  pie=3.14
  def __init__(self,radius,height):
    self.radius=radius
    self.height=height

  def getVolume(self):
      return self.pie * self.radius * self.radius * (self.height/3)

  def getTotalSurfaceArea(self):
    return self.pie * self.radius * (self.radius + (math.sqrt (math.pow(self.height,2)+math.pow(self.radius,2) ) )  )

  def getCurvedSurfaceArea(self):
    return self.pie * self.radius *  (math.sqrt (math.pow(self.height,2)+math.pow(self.radius,2) ) ) 
```

```
c1= cone(5,12)
print("Total Surface Area:",c1.getTotalSurfaceArea())
print("Curved Surface Area:",c1.getCurvedSurfaceArea())
print("Volume:",c1.getVolume())
```
