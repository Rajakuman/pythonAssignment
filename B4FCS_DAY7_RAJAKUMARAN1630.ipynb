**Question 1:**
For this challenge you need to develop a Python program to open a file in read only mode and try writing something in it and handle the subsequent errors using Exception Handling.


```
try: # We are trying a peice of code and lets see if it break or not
    print("we are in try block")
    file = open("rajaMain.txt",'r')
    statement = file.read()
    print(statement)
    file.write("Write somthing in read mode file")
    print(statement)
except Exception as e: # Catch the error and print something to conclude 
    print("There was an error - ", e)
    
finally: # Finally is going to run what ever may happen 
    print("We are in Finally block")
    file.close()
```

**Question 2:**
  Write a python Function for finding a given number prime or not and do Unit Testing on it using PyLint & Unit test Library


```
%%writefile isprimary.py
''' 
Program to check if a number is prime or not based on input
'''
INP=17
# inp = int(input("Enter a number: "))
# prime numbers are greater than 1
def is_prime_number(num):
    '''
    function for found is prime number
    '''
    bol=False
    if num>2:
      for i in range(2,num):
        if (num % i) == 0:
          print(num,"is not a prime number")
          print(i,"times",num//i,"is",num)
          break
        else:
          bol=True
      if bol:
        print(num,"is a prime number")
# if input number is less than
# or equal to 1, it is not prime
    else:
      print(num,"is not a prime number")

is_prime_number(INP)
```
