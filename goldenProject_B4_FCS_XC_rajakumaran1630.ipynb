Akash and Vishal are quite fond of travelling. 

They mostly travel by railways. 

They were travelling in a train one day and they got interested in the seating arrangement of their compartment. 

The compartment looked something like - https://letsupgrade.in/projectIMages/train.jpg


So they got interested to know the seat number facing them and the seat type facing them. The seats are denoted as follows :

Window Seat : WS

Middle Seat : MS

Aisle Seat : AS

You will be given a seat number, find out the seat number facing you and the seat type, i.e. WS, MS or AS.


----------------------------------
INPUT

First line of input will consist of a single integer T denoting number of test-cases. Each test-case consists of a single integer N denoting the seat-number.
----------------------------------
----------------------------------
OUTPUT

For each test case, print the facing seat-number and the seat-type, separated by a single space in a new line.
----------------------------------
CONSTRAINTS

1<=T<=105

1<=N<=108

****************
SAMPLE INPUT
****************
2

18

40


****************
SAMPLE OUTPUT
****************
19 WS
45 AS



```
no_of_seats=int( input("Enter no of Trian seat:") )
lst = []
if (1<=no_of_seats and no_of_seats<=105):
    for lp in range(no_of_seats):
        input_seatnumbers=int(input())
        lst.append(input_seatnumbers)
else:
    print("Enter proper input, Values are 1 to 105")

# print(lst)
def getseat(n):
    if n>=1 and n<=108:
        if n%6==1 or  n%6==0  :
            return "WS"
        elif n%6==2 or  n%6==5  :
            return "MS"
        elif n%6==3 or  n%6==4  :
            return "AS"
    else:
        return "Seat number is invalid!"

if len(lst) != 0:
    print("output:------------------------------------------")
    for ll in lst:
        print(ll,getseat(ll))
    print("-------------------------------------------------")
```
